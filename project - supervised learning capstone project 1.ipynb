Broader Cohort Groups


Based on the tenure column values,create a new column called Tenure Cohort that creates 4 separate categories

```python
df['tenure']
```

```python
def Cohort (tenure):
    if tenure<13:
        return '0-12 months'
    elif 12<tenure<25:
        return "13-24 months"
    elif 24<tenure<49:
        return "25-48 months"
    else:
        return("over 48 months")
```

```python
df["tenure"].apply(Cohort)
```

```python
df['Tenure Cohort']=df["tenure"].apply(Cohort)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x="MonthlyCharges",y="TotalCharges",hue="Tenure Cohort",alpha=0.4,linewidth=0.3)
```

# 1)Single Decision Tree

TASK : Separate out the data into X features and Y label. Create dummy variables where necessary and note which features are not useful and should be dropped

```python
df
```

```python
X = df.drop(["Churn","customerID"],axis=1)
X = pd.get_dummies(X,drop_first=True)
y = df['Churn']
```

TASK : Perform train_test_split, holding out 10% of the data for testing.We will use a random_state 101 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report , plot_confusion_matrix ,accuracy_score
```

```python
plot_confusion_matrix(dt,X_test,y_test)
```

```python
print(classification_report(y_test,preds))
```
